##Exercício 01 - Filtrar a data mais recente (realizar o processo para buscar a última data de forma automatizada) e os países Brasil, Itália, E.U.A e Canadá.

```python
#Filtro automatizado data recente
max_date = np.max(dados_covid['date'])


dataMaisRecente = dados_covid[(dados_covid['date'] == max_date)]

filtro_paises = dataMaisRecente.groupby('country').agg(mortes = pd.NamedAgg(column = 'dead', aggfunc = np.sum), 
                                                       infectados = pd.NamedAgg(column = 'infected',aggfunc = np.sum),) \
                                                       .sort_values('mortes').reset_index()

                                                  


filtro_paises = filtro_paises[filtro_paises['country'].isin(['Brazil', 'US', 'Italy','Canada'])]


filtro_paises
```

```python
#Filtro data espeficica
dataMaisAntinga = dados_covid[(dados_covid['date'] == '2020-04-18')]

filtro_paises_data_antiga = dataMaisAntinga.groupby('country').agg(mortes = pd.NamedAgg(column = 'dead', aggfunc = np.sum), 
                                                       infectados = pd.NamedAgg(column = 'infected',aggfunc = np.sum),) \
                                                       .sort_values('mortes').reset_index()     

filtro_paises_data_antiga = filtro_paises_data_antiga[filtro_paises_data_antiga['country'].isin(['Brazil', 'US', 'Italy','Canada'])]

filtro_paises_data_antiga
```

###**Comentarios em relação aos dados:**

#####Comparando a evolução da pandemia entre os países, observamos que a situação do Brasil é similar com a dos Estados Unidos e diferente da Itália. Enquanto a Itália, viu a quantidade de casos aumentar de forma drástica quando atingiu o pico da pandemia em 2020, o governo do país tomou medidas para impedir o avanço da Covid-19 e isso fez com que o país saisse da primeira posição de infectados e mortes dado o numero da sua população em 2020s. No Brasil e nos EUA não houve até o momento uma regressão geral e consistente da pandemia.
#####Podemos considerar alguns fatores que contribuíram para esse comportamento. Em primeiro lugar seria a dimensão continental do Brasil e dos Estados Unidos. Ambos os países são considerados as maiores nações do mundo em extensão territorial, e a Itália é quase do tamanho do Maranhão. A expansão geográfica influencia diretamente como cada país enfrenta a epidemia, pois cada região pode estar em uma fase diferente da pandemia e isso afeta a curva de infectados e óbitos em cada país.
- Estados Unidos
#####Analisando as tabelas referente a probabilidade de infectados X probabilidade de mortos, podemos assumir que os **EUA tem a maior taxa de infectados**, no entanto, o **número de mortes é menor**. Uma hipótese que pode explicar esta situação é o sistema de saúde ser mais eficaz e organizado em comparação ao Brasil e Itália, além da campanha de vacinação ter sido iniciado em 2020.
- Itália 
##### A Itália foi o primeiro país a implementar a quarentena a nível nacional, durante o ano de 2020 houve 3 lockdowns para tentar freiar a quantidade do número de infectados no país. No entanto, observamos na tabela que mesmo com as medidas tomadas pelo governo a probabilidade de infectados da Itália comparado a quantidade da população mundial e com a população local aumentou entre os anos de 2020 e 2021. Sendo o país com **maior probabilidade de mortes** comparando sua população. **Uma hipotese para isso, é que uma grande parcela da sua população está no grupo de risco devido idade avançada.**
- Brasil
##### O caráter negacionista do governo brasileiro perante a Pandemia fez com que a contabilização de números de infectados no ano de 2020 fosse ineficiente. Além disso, o número de testagem contra a COVID 19 no país é baixa, o que refletiu em uma baixa probabilidade de infecção em 2020 quando comparamos a população dos três países analisados. Quando analisamos as probabilidades somente em relação aos brasileiros, verificamos que em 2020 a probabilidade de infectados é expressivamente menor, o que reforça a análise anterior. 
###### Outro ponto de destaque é o grande aumento na probabilidade de infectados no Brasil em 2021. Como pontos corroboraram para esse cenário, além daqueles citados anteriormente, temos a ineficiência em uma política federal quanto à realizações de lockdown, bem como a pauta do Auxílio Emergencial que possibilita que ele seja feito.
